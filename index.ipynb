Use the function created in Exercise 1 to answer the following analytical questions.

In a Python class, some of the students have taken online courses to improve their Python skills.
The scores of a random sample of 20 students who underwent the online-course on a Python test are as follows:

      78.7, 73.5, 86.9, 87.4, 82.7, 81.9, 69.9, 77.2, 79.3, 83.3]

The mean score of the class before the test is 65. The teacher thinks that the online training has really helped the students and now they should perform better than the class (population) mean. Use this to set your null and alternative hypotheses.

1. Test to see if the sample mean is significantly different from 65 at the .05 level. Report the t- and p-values.
2. The researcher realizes that she accidentally recorded the score that should have been 80.9 as 90.9. Are these corrected scores significantly different from 65 at the .05 level?

Bonus: What is the effect size of the first sample compared to the population? How can you interpret this effect size in relation to your significance test?


### Solution:

```python
import numpy as np
import scipy.stats as stats
import seaborn as sns
import matplotlib.pyplot as plt

def one_sample_ttest(sample, popmean, alpha):
    # Visualize sample distribution for normality
    sns.histplot(sample, kde=True)
    plt.title('Sample Distribution')
    plt.xlabel('Value')
    plt.ylabel('Frequency')
    plt.show()
    
    # Sample mean (x̄) using NumPy mean()
    sample_mean = np.mean(sample)
    
    # Sample Standard Deviation (sigma) using NumPy std()
    sample_std = np.std(sample, ddof=1)  # Using ddof=1 to get the sample standard deviation
    
    # Sample size
    n = len(sample)
    
    # Degrees of freedom
    df = n - 1
    
    # Calculate the t-statistic
    t_stat = (sample_mean - popmean) / (sample_std / np.sqrt(n))
    
    # Calculate the critical t-value for the given alpha
    critical_t = stats.t.ppf(1 - alpha/2, df)
    
    # Calculate the p-value
    p_value = (1 - stats.t.cdf(abs(t_stat), df)) * 2
    
    # Determine if the results are significant
    if abs(t_stat) > critical_t:
        result = "The result is significant. Reject the null hypothesis."
    else:
        result = "The result is not significant. Fail to reject the null hypothesis."
    
    # Print results
    print(f"Sample mean: {sample_mean}")
    print(f"Sample standard deviation: {sample_std}")
    print(f"t-statistic: {t_stat}")
    print(f"Critical t-value: {critical_t}")
    print(f"p-value: {p_value}")
    print(result)

    return t_stat, critical_t, p_value, result

def cohen_d(sample, popmean):
    sample_mean = np.mean(sample)
    sample_std = np.std(sample, ddof=1)
    return (sample_mean - popmean) / sample_std

# Sample scores
sample_scores = [84.0, 92.4, 74.3, 79.4, 86.7, 75.3, 90.9, 86.1, 81.0, 85.1, 
                 78.7, 73.5, 86.9, 87.4, 82.7, 81.9, 69.9, 77.2, 79.3, 83.3]

# Population mean
pop_mean = 65

# Significance level
alpha = 0.05

# Perform one-sample t-test on original scores
print("Original Scores:")
t_stat, critical_t, p_value, result = one_sample_ttest(sample_scores, pop_mean, alpha)

# Corrected sample scores
corrected_sample_scores = [84.0, 92.4, 74.3, 79.4, 86.7, 75.3, 80.9, 86.1, 81.0, 85.1, 
                           78.7, 73.5, 86.9, 87.4, 82.7, 81.9, 69.9, 77.2, 79.3, 83.3]

# Perform one-sample t-test with corrected scores
print("\nCorrected Scores:")
corrected_t_stat, corrected_critical_t, corrected_p_value, corrected_result = one_sample_ttest(corrected_sample_scores, pop_mean, alpha)

# Calculate Cohen's d for the original sample
effect_size = cohen_d(sample_scores, pop_mean)
print(f"\nEffect size (Cohen's d) for the original sample: {effect_size}")

# Calculate Cohen's d for the corrected sample
corrected_effect_size = cohen_d(corrected_sample_scores, pop_mean)
print(f"Effect size (Cohen's d) for the corrected sample: {corrected_effect_size}")
```
